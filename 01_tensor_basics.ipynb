## 1. Tensor란 무엇인가?

Tensor는 PyTorch에서 사용하는 다차원 배열 자료형이다.  
수치 데이터를 저장하고 연산하기 위한 기본 단위이며, 딥러닝 모델의 입력, 가중치, 출력 등을 표현하는 데 사용된다.

## 2. Tensor와 NumPy의 차이

NumPy 배열은 주로 CPU 기반의 수치 연산에 사용되며, 자동 미분 기능을 지원하지 않는다.  
반면 PyTorch tensor는 NumPy와 유사한 배열 구조를 가지지만, GPU 연산과 자동 미분을 지원하여 딥러닝 학습에 더 적합하다.

## 3. 딥러닝에서 Tensor를 사용하는 이유

딥러닝에서는 입력 데이터, 가중치, 출력값을 모두 다차원 형태로 표현해야 하므로 tensor를 사용한다.  
또한 tensor는 자동 미분과 GPU 연산을 지원하기 때문에, 모델 학습과 역전파를 보다 효율적으로 수행할 수 있다.

```python
import torch
import numpy as np
```

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
```

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

```python
x_ones = torch.ones_like(x_data)
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float)
print(f"Random Tensor: \n {x_rand} \n")
```

```python
shape = (2,3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor} \n")
```

```python
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
tensor = torch.ones(4, 4)
print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")
tensor[:,1] = 0
print(tensor)
```

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 =  torch.rand_like(y1)
torch.matmul(tensor, tensor.T, out=y3)

z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
```

```python
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
```

```python
print(f"{tensor} \n")
tensor.add_(5)
print(tensor)
```

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

```python
n = np.ones(5)
t = torch.from_numpy(n)
```

```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
